```python
#Import Dependencies

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import CSV

training_data = pd.read_csv('Bitcoin_Stock_Price_Train.csv')

training_data.shape
training_data.head()
```

```python
training_data = training_data.iloc[:, 1:2]

training_data.shape
training_data.head()
```

```python
#Feature Scaling

from sklearn.preprocessing import MinMaxScaler

mm = MinMaxScaler(feature_range = (0, 1))
training_data = mm.fit_transform(training_data)
```

```python
#Getting the Inputs and Outputs

x_train = training_data[0:1257]
y_train = training_data[1:1258]

print(x_train.shape)
print(y_train.shape)
```

```python
#Reshaping

x_train = np.reshape(x_train, (1257, 1, 1))

print(x_train.shape)
```

```python
#Importing the Keras Libraries and Packages

import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
```

```python
#Initializing the Model
model = Sequential()

#Adding the Input Layer and the LSTM Layer
model.add(LSTM(units = 4, activation = 'sigmoid', input_shape = (None, 1)))

#Adding the Output Layer
model.add(Dense(units = 1))

#Compiling the Model
model.compile(optimizer = 'adam', loss = 'mean_squared_error')

#Fitting the RNN to the Training Data
model.fit(x_train, y_train, batch_size = 32, epochs = 200)
```

```python
# getting the real stock of 2020 (June-December) i.e., importing the test dataset

test_data = pd.read_csv('Bitcoin_Stock_Price_Test.csv')
real_stock_price = test_data.iloc[:,1:2]
real_stock_price.head()
```

```python
#Getting the Predicted Stock Price of 2020 (June-December)

inputs = real_stock_price
inputs = mm.transform(inputs)
inputs = np.reshape(inputs, (214, 1, 1))


predicted_stock_price = model.predict(inputs)
predicted_stock_price = mm.inverse_transform(predicted_stock_price)
```

```python
predicted_stock_price
pd.DataFrame(predicted_stock_price).to_csv("predicted_stock_price.csv")
```

```python
#Visualizing the Results

plt.plot(real_stock_price, color = 'red', label = 'Real Bitcoin Stock Price')
plt.plot(predicted_stock_price, color = 'blue', label = 'Predicted Bitcoin Stock Price')
plt.title('Stock Prediction')
plt.xlabel('Time')
plt.ylabel('Bitcoin Stock Price')
plt.legend()
plt.show()
```
